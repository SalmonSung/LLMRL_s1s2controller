```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

folder1 = 'data'
folder2 = 'data2'

files1 = set(f for f in os.listdir(folder1) if f.endswith('.csv'))
files2 = set(f for f in os.listdir(folder2) if f.endswith('.csv'))

# 取两个文件夹的交集文件名
common_files = sorted(list(files1 & files2))

ratios1 = []
ratios2 = []

for fname in common_files:
    df1 = pd.read_csv(os.path.join(folder1, fname))
    df2 = pd.read_csv(os.path.join(folder2, fname))

    ratio1 = df1['pass_or_fail'].mean()  # pass_or_fail 是0/1，mean即为比例
    ratio2 = df2['pass_or_fail'].mean()

    ratios1.append(ratio1)
    ratios2.append(ratio2)

x = np.arange(len(common_files))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, ratios1, width, label=folder1)
bars2 = ax.bar(x + width/2, ratios2, width, label=folder2)

ax.set_xlabel('文件名')
ax.set_ylabel('pass_or_fail=1 比例')
ax.set_title('两个文件夹同名文件中 pass_or_fail=1 的比例对比')
ax.set_xticks(x)
ax.set_xticklabels(common_files, rotation=45, ha='right')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读data.csv
df = pd.read_csv('data.csv')

# 统计0和1的数量
counts = df['s1_or_s2'].value_counts().sort_index()  # 确保0和1顺序

labels = ['s1 (0)', 's2 (1)']
values = [counts.get(0, 0), counts.get(1, 0)]

# 计算比例
total = sum(values)
ratios = [v / total for v in values]

# 画柱状图
plt.figure(figsize=(6,4))
bars = plt.bar(labels, ratios, color=['blue', 'orange'])
plt.ylim(0,1)
plt.ylabel('比例')
plt.title('data.csv 中 s1_or_s2 标签比例')

# 在柱上标注具体比例
for bar, ratio in zip(bars, ratios):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
             f'{ratio:.2%}', ha='center', fontsize=12)

plt.show()
```
